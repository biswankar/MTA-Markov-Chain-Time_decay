```python
def matrix_transform(conv_act):
    conv_act = np.fromstring(conv_act,float).reshape(5,6)
    start_idx = conv_act[-1].copy()
    for i in range(0,conv_act.shape[1]):
        conv_act[:,i] = conv_act[:,i]/(conv_act[:,i].sum())
    conv_act[-1] = start_idx/start_idx.sum()
    return(conv_act.tostring())

#cust_master['transition_matrix'].apply(lambda row : matrix_transform(row)) 
#cust_master
cust_master['prob_matrix'] = cust_master['transition_matrix'].apply(matrix_transform)
cust_master.head() 
```

```python
txn_df = newdf[newdf.txn_break!=0]
txn_df['phone'] = txn_df.phone.astype('str')
txn_df['date'] = pd.to_datetime(txn_df['date']).dt.date

#txn_df = txn_df.head(13)
act_list = ['SMS', 'email', 'App', 'WEB' ]
for i in range(0,len(act_list)):
        cust_master[str(act_list[i])+'_rem_effect'] = 0
cust_master['txn_prob'] = 0

from_axis = {'SMS': 0, 'email': 1, 'App': 2, 'WEB':3 , 'start' : 4}
to_axis = {'SMS': 0, 'email': 1, 'App': 2, 'WEB':3 , 'Txn' : 4}

for index,row in cust_master.iterrows():
    temp_df = txn_df[txn_df.phone == row['customer']]
    #temp_df = temp_df.sort_values(by = ['phone','date'])
    temp_df.sort_values(by = ['phone','date'], inplace=True)
    
    total_prob = '0'
    start = 1
    
    trans_mat = np.fromstring(row['prob_matrix'],float).reshape(5,6)
    #print(trans_mat)
    temp_df.sort_values(by = ['phone','date'], inplace=True)
    #print(temp_df.sort_values(by = ['phone','date'], inplace=True))
    #print(temp_df)
    for i in range(0,temp_df.shape[0]-1):
        #print(temp_df)
        #print(temp_df.iloc[i,:])
        #print(temp_df.iloc[i,3]+temp_df.iloc[i+1,3])
        if(temp_df.iloc[i,3] == 'Txn'):
            start = 1
            continue
        if(start == 1):
            if(temp_df.iloc[i,3] == 'Txn'):
                continue
            else:
                total_prob = total_prob + '+' +str(trans_mat[4][to_axis[temp_df.iloc[i,3]]])+'*'+str(trans_mat[from_axis[temp_df.iloc[i,3]]][to_axis[temp_df.iloc[i+1,3]]])  
                start = 0
                continue
        total_prob = total_prob +'*'+str(trans_mat[from_axis[temp_df.iloc[i,3]]][to_axis[temp_df.iloc[i+1,3]]])  
    txn_prob = eval(total_prob)
    rem_effect = []
    
    for ac in act_list:
        rem_df = temp_df[~temp_df.txn_id.isin(temp_df[temp_df.act == str(ac)]['txn_id'].unique())]
        start = 1
        rem_prob = '0'
        rem_df.sort_values(by = ['phone','date'],ascending = (True,True),inplace = True)
        for i in range(0,rem_df.shape[0]-1):
            if(rem_df.iloc[i,3] == 'Txn'):
                start = 1
                continue
            if(start == 1):
                if(rem_df.iloc[i,3] == 'Txn'):
                    continue
                else:
                    rem_prob = rem_prob + '+' +str(trans_mat[4][to_axis[rem_df.iloc[i,3]]])+'*'+str(trans_mat[from_axis[rem_df.iloc[i,3]]][to_axis[rem_df.iloc[i+1,3]]])  
                    start = 0
                    continue
            #print(i)
            #print()
            #print(to_axis[rem_df.iloc[i+1,3]])
            
            rem_prob = rem_prob +'*'+str(trans_mat[from_axis[rem_df.iloc[i,3]]][to_axis[rem_df.iloc[i+1,3]]])  
        txn_rem_prob = eval(rem_prob)
        rem_effect.append(txn_rem_prob)
        #rem_df.info()
        #print(rem_df)
        del rem_df
        del rem_prob
    del temp_df
    del total_prob
    for i in range(0,len(act_list)):
        cust_master.loc[index,str(act_list[i])+'_rem_effect'] = rem_effect[i]
    cust_master.loc[index,'txn_prob'] = txn_prob
## Do the 1-total_prob or act/total_prob after confimation 
```
